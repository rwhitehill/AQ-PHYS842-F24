```python
import numpy as np
from scipy.integrate import quad

import matplotlib.pyplot as plt
plt.rcParams.update({
    'text.latex.preamble': r'\usepackage{amsmath}',
    'text.usetex': True,
    'font.family': 'sans-serif',
    'font.sans-serif': ['Helvetica']
})
```

```python
def I(a):
    integrand = lambda x,a: x**2*np.exp(-x**2/2)/np.sqrt(x**2 + a**2)
    return quad(integrand,0,100,args=(a))[0]
```

```python
a = np.logspace(-5,5,100)
I_ = np.array([I(_) for _ in a])
```

```python
nrows,ncols = 1,1
fig,ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,5*nrows))

ax.plot(a,I_)
ax.semilogx()
ax.tick_params(axis='both',which='major',labelsize=20,direction='in')
ax.set_xlabel(r'$m \ell$',size=30)
ax.set_ylabel(r'$(2 \pi \ell)^2 \sigma^2$',size=30)

plt.show()
fig.savefig('prob4.pdf',bbox_inches='tight')
```
